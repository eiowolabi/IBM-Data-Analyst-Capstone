```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp FROM master
"""

df = pd.read_sql_query(QUERY,conn)
print(df.head())

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
%matplotlib inline

plt.figure(figsize=(10,5))
sns.distplot(a=df,bins=20,kde=False)
plt.xlabel('Salary')
plt.ylabel('Respondents')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age FROM master
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

df['Age'].plot(kind='box', figsize=(8, 6))

plt.title('Age of Survey Respondents')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')

plt.title('Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

```python
QUERY = """
SELECT Age, ConvertedComp FROM master
where Age between 25 and 30
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

age = df.groupby("Age").median()[["ConvertedComp"]]
age.plot(kind='line')

plt.title('Median ConvertedComp for Age 25 - 30')
plt.ylabel('Median ConvertedComp')
plt.xlabel('Age')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x="CodeRevHrs", y="WorkWeekHrs", size="Age", legend=False, sizes=(20, 2000))
plt.title('CodeRevHrs vs WorkWeekHrs (Weight by Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
#pd.set_option('display.max_columns', None)
QUERY = """
SELECT DatabaseDesireNextYear, count(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc limit 5
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

x = df['count']
label = df['DatabaseDesireNextYear']
plt.pie(x, labels=label, autopct='%1.1f%%',radius=1.5)
```

```python
#LanguageDesireNextYear
QUERY = """
SELECT LanguageDesireNextYear, count(*) as count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count desc limit 5
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

x = df['count']
label = df['LanguageDesireNextYear']
plt.pie(x, labels=label, autopct='%1.1f%%',radius=1.5)
```

```python
#LanguageWorkedWith
QUERY = """
SELECT LanguageWorkedWith, count(*) as count
FROM LanguageWorkedWith
group by LanguageWorkedWith
order by count desc
"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
#DatabaseWorkedWith
QUERY = """
SELECT Respondent, count(Respondent)
FROM DatabaseWorkedWith
where Respondent in (select Respondent from DatabaseWorkedWith where DatabaseWorkedWith = 'MySQL')
group by Respondent
Having count(Respondent) = 1
"""
df = pd.read_sql_query(QUERY,conn)
print(df.shape)
```

```python
#DevType
QUERY = """
SELECT DevType, count(*) as count
FROM DevType
group by DevType
order by count desc
"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
pd.set_option('display.max_rows', None)
QUERY = """
With WWH as 
(select WorkWeekHrs, ROW_NUMBER() OVER () AS Ind FROM (select WorkWeekHrs from master where Age between 30 and 35 and WorkWeekHrs is not null
order by WorkWeekHrs) a ),
CRH as 
(select CodeRevHrs, ROW_NUMBER() OVER () AS Ind FROM (select CodeRevHrs from master where Age between 30 and 35 and CodeRevHrs is not null
order by CodeRevHrs) a )
Select W.WorkWeekHrs, C.CodeRevHrs, '30 - 35' as Age_Group from WWH W, CRH C
where W.Ind = (select (count(*) + 1) / 2 from WWH)
and C.Ind = (select (count(*) + 1) / 2 from CRH)

"""
df = pd.read_sql_query(QUERY,conn)
print(df)

# plot bars in stack manner
df.plot(x='Age_Group', kind='bar', stacked=True,
        title='Stacked Bar Graph by dataframe')
#plt.bar(df['Median_WrkWkHrs'], df[Median_CodeRevHrs], color='r')
#plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT Age, --string_list, count, middle,
    CASE WHEN count%2=0 THEN
        0.5 * substr(string_list, middle-10, 10) + 0.5 * substr(string_list, middle, 10)
    ELSE
        1.0 * substr(string_list, middle, 10)
    END AS median
FROM (
    SELECT Age, 
        group_concat(ConvertedComp_string,"") AS string_list,
        count() AS count, 
        1 + 10*(count()/2) AS middle
    FROM (
        SELECT Age, 
            printf( '%010d',ConvertedComp) AS ConvertedComp_string
        FROM master
        ORDER BY Age,ConvertedComp_string
    )
    GROUP BY Age
)
where Age between 45 and 60
"""
df = pd.read_sql_query(QUERY,conn)
print(df)

age = df.groupby("Age").mean()[["median"]]
age.plot(kind='line')

plt.title('Median ConvertedComp for Age 45 - 60')
plt.ylabel('Median')
plt.xlabel('Age')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, count(*) as count
FROM master
group by MainBranch
"""
df = pd.read_sql_query(QUERY,conn)
print(df.head())

df_bar = df.groupby("MainBranch").mean()[['count']]
df_bar.plot(kind='barh', figsize=(12, 12), color='steelblue')

plt.xlabel('Main Branch')
plt.title('Main Branch of Survey Respondents')

plt.show()
```
